```python
# ================================================================
# InternBridge: Data Science Internship Skill Gap Analyzer
# Full Beginner-Friendly Analysis Code
# Dataset: all_job_post.csv / all_job_post.xlsx
# Author: Chamodi Kumarage
# ================================================================


# ================================================================
# 1. IMPORT REQUIRED LIBRARIES
# ================================================================

# pandas is used for reading, cleaning, and analyzing data
import pandas as pd

# numpy is used for numerical calculations
import numpy as np

# matplotlib is used for creating charts
import matplotlib.pyplot as plt

# ast is used to safely convert text lists into real Python lists
import ast

# re is used for text cleaning
import re

# pathlib is used to handle file and folder paths
from pathlib import Path

# Counter is used to count skill frequencies
from collections import Counter

# TF-IDF is used for basic NLP keyword extraction
from sklearn.feature_extraction.text import TfidfVectorizer

# Cosine similarity is used for matching student skills with job skills
from sklearn.metrics.pairwise import cosine_similarity


# ================================================================
# 2. SET PROJECT FOLDERS
# ================================================================

# This code assumes your notebook is inside:
# D:\Projects\InternBridge\notebooks
#
# If you opened Jupyter from D:\Projects\InternBridge,
# this code will still work.

current_folder = Path.cwd()

if current_folder.name.lower() == "notebooks":
    project_folder = current_folder.parent
else:
    project_folder = current_folder

data_folder = project_folder / "data"
outputs_folder = project_folder / "outputs"
charts_folder = project_folder / "charts"
reports_folder = project_folder / "reports"

# Create folders automatically if they do not exist
data_folder.mkdir(exist_ok=True)
outputs_folder.mkdir(exist_ok=True)
charts_folder.mkdir(exist_ok=True)
reports_folder.mkdir(exist_ok=True)

print("Project folder:", project_folder)
print("Data folder:", data_folder)
print("Outputs folder:", outputs_folder)
print("Charts folder:", charts_folder)
print("Reports folder:", reports_folder)


# ================================================================
# 3. LOAD THE DATASET
# ================================================================

# Expected file location:
# D:\Projects\InternBridge\data\all_job_post.csv
#
# If your file is Excel, save it as:
# D:\Projects\InternBridge\data\all_job_post.xlsx

csv_file = data_folder / "all_job_post.csv"
excel_file = data_folder / "all_job_post.xlsx"

if csv_file.exists():
    df = pd.read_csv(csv_file)
    print("CSV file loaded successfully.")
elif excel_file.exists():
    df = pd.read_excel(excel_file)
    print("Excel file loaded successfully.")
else:
    raise FileNotFoundError(
        "Dataset not found. Please place all_job_post.csv or all_job_post.xlsx inside the data folder."
    )

# Display first 5 rows
df.head()


# ================================================================
# 4. BASIC DATASET INSPECTION
# ================================================================

print("Dataset shape:")
print(df.shape)

print("\nColumn names:")
print(df.columns.tolist())

print("\nDataset information:")
df.info()

print("\nFirst 5 rows:")
display(df.head())

print("\nLast 5 rows:")
display(df.tail())


# ================================================================
# 5. CLEAN COLUMN NAMES
# ================================================================

# Convert column names to lowercase
# Remove extra spaces
# Replace spaces with underscores

df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

print("Cleaned column names:")
print(df.columns.tolist())


# ================================================================
# 6. CHECK MISSING VALUES AND DUPLICATES
# ================================================================

print("Missing values in each column:")
print(df.isnull().sum())

print("\nNumber of duplicate rows:")
print(df.duplicated().sum())

# Remove duplicate rows if any exist
df = df.drop_duplicates()

print("\nDataset shape after removing duplicates:")
print(df.shape)


# ================================================================
# 7. CHECK EXPECTED COLUMNS
# ================================================================

# These are the columns expected in your dataset
expected_columns = [
    "job_id",
    "category",
    "job_title",
    "job_description",
    "job_skill_set"
]

missing_columns = [col for col in expected_columns if col not in df.columns]

if len(missing_columns) > 0:
    print("Warning: These expected columns are missing:")
    print(missing_columns)
else:
    print("All expected columns are available.")


# ================================================================
# 8. CLEAN TEXT DATA
# ================================================================

# Convert all important columns to text
# Remove extra spaces
# Standardize category names

df["job_id"] = df["job_id"].astype(str).str.strip()
df["category"] = df["category"].astype(str).str.strip().str.upper()
df["job_title"] = df["job_title"].astype(str).str.strip()
df["job_description"] = df["job_description"].astype(str).str.strip()
df["job_skill_set"] = df["job_skill_set"].astype(str).str.strip()

print("Text cleaning completed.")
display(df.head())


# ================================================================
# 9. CONVERT SKILL TEXT INTO PYTHON LISTS
# ================================================================

# In the dataset, skills may look like this:
# ['Python', 'SQL', 'Excel']
#
# Python reads it as text first.
# This function converts it into a real Python list.

def convert_skills(skill_text):
    """
    Convert the job_skill_set text into a clean Python list.
    This function handles different possible formats safely.
    """
    try:
        skills = ast.literal_eval(skill_text)
        
        if isinstance(skills, list):
            return [str(skill).strip().lower() for skill in skills if str(skill).strip() != ""]
        else:
            return []
    
    except:
        # Backup method if ast conversion fails
        skill_text = str(skill_text).replace("[", "").replace("]", "").replace("'", "").replace('"', "")
        skills = re.split(r",|;|\|", skill_text)
        return [skill.strip().lower() for skill in skills if skill.strip() != ""]


df["skills_list"] = df["job_skill_set"].apply(convert_skills)

print("Skill list conversion completed.")
display(df[["job_title", "job_skill_set", "skills_list"]].head())


# ================================================================
# 10. CREATE USEFUL NEW COLUMNS
# ================================================================

# Count how many skills are required for each job post
df["skill_count"] = df["skills_list"].apply(len)

# Count how many words are in each job description
df["description_word_count"] = df["job_description"].apply(lambda x: len(str(x).split()))

# Create a combined text column for NLP analysis
df["combined_text"] = (
    df["job_title"].astype(str) + " " +
    df["job_description"].astype(str) + " " +
    df["job_skill_set"].astype(str)
)

print("New columns created successfully.")
display(df[["job_title", "skill_count", "description_word_count"]].head())


# ================================================================
# 11. BASIC DESCRIPTIVE ANALYSIS
# ================================================================

print("Total number of job posts:", len(df))
print("Total number of job categories:", df["category"].nunique())
print("Total number of unique job titles:", df["job_title"].nunique())

print("\nJob count by category:")
category_count = df["category"].value_counts()
display(category_count)

print("\nSummary statistics for skill count:")
display(df["skill_count"].describe())

print("\nSummary statistics for job description word count:")
display(df["description_word_count"].describe())


# ================================================================
# 12. CATEGORY-WISE ANALYSIS
# ================================================================

# Average number of skills required by each category
avg_skills_by_category = (
    df.groupby("category")["skill_count"]
    .mean()
    .sort_values(ascending=False)
)

print("Average number of skills required by category:")
display(avg_skills_by_category)

# Average description length by category
avg_description_by_category = (
    df.groupby("category")["description_word_count"]
    .mean()
    .sort_values(ascending=False)
)

print("Average job description length by category:")
display(avg_description_by_category)

# Number of job posts and average skills together
category_summary = df.groupby("category").agg(
    number_of_jobs=("job_id", "count"),
    average_skill_count=("skill_count", "mean"),
    average_description_words=("description_word_count", "mean")
).sort_values(by="number_of_jobs", ascending=False)

print("Category summary:")
display(category_summary)


# ================================================================
# 13. SKILL FREQUENCY ANALYSIS
# ================================================================

# Explode means one skill becomes one row
skills_exploded = df.explode("skills_list")

# Rename the exploded skill column for clarity
skills_exploded = skills_exploded.rename(columns={"skills_list": "skill"})

# Clean skill names again
skills_exploded["skill"] = skills_exploded["skill"].astype(str).str.strip().str.lower()

# Remove empty skills
skills_exploded = skills_exploded[skills_exploded["skill"] != ""]

# Count top skills
top_skills = skills_exploded["skill"].value_counts().head(20)

print("Top 20 most required skills:")
display(top_skills)


# ================================================================
# 14. TOP SKILLS BY EACH CATEGORY
# ================================================================

print("Top 10 skills by category:")

top_skills_by_category = {}

for category in df["category"].unique():
    category_data = skills_exploded[skills_exploded["category"] == category]
    top_10 = category_data["skill"].value_counts().head(10)
    top_skills_by_category[category] = top_10
    
    print("\nCategory:", category)
    display(top_10)


# ================================================================
# 15. FILTER DATA-RELATED JOBS
# ================================================================

# These keywords are related to Data Science, Data Analysis, BI, AI, and ML
data_keywords = (
    "data|analyst|analytics|business intelligence|bi|"
    "machine learning|ml|artificial intelligence|ai|"
    "scientist|python|sql|database|reporting|dashboard|"
    "power bi|tableau|statistics"
)

data_jobs = df[
    df["job_title"].str.lower().str.contains(data_keywords, na=False) |
    df["job_description"].str.lower().str.contains(data_keywords, na=False) |
    df["job_skill_set"].str.lower().str.contains(data_keywords, na=False)
].copy()

print("Number of data-related job posts:", len(data_jobs))
print("Percentage of data-related jobs:", round((len(data_jobs) / len(df)) * 100, 2), "%")

display(data_jobs[["job_id", "category", "job_title", "skill_count"]].head(20))


# ================================================================
# 16. ANALYZE SKILLS IN DATA-RELATED JOBS
# ================================================================

data_skills = data_jobs.explode("skills_list")
data_skills = data_skills.rename(columns={"skills_list": "skill"})
data_skills["skill"] = data_skills["skill"].astype(str).str.strip().str.lower()
data_skills = data_skills[data_skills["skill"] != ""]

top_data_skills = data_skills["skill"].value_counts().head(20)

print("Top 20 skills required for data-related jobs:")
display(top_data_skills)


# ================================================================
# 17. IT / INFORMATION TECHNOLOGY JOB ANALYSIS
# ================================================================

it_jobs = df[df["category"].str.contains("INFORMATION", na=False)].copy()

print("Number of Information Technology job posts:", len(it_jobs))

it_skills = it_jobs.explode("skills_list")
it_skills = it_skills.rename(columns={"skills_list": "skill"})
it_skills["skill"] = it_skills["skill"].astype(str).str.strip().str.lower()
it_skills = it_skills[it_skills["skill"] != ""]

top_it_skills = it_skills["skill"].value_counts().head(20)

print("Top 20 skills required for Information Technology jobs:")
display(top_it_skills)


# ================================================================
# 18. BASIC NLP ANALYSIS USING TF-IDF
# ================================================================

# TF-IDF helps identify important words in job descriptions.
# It gives higher importance to words that appear often in one document
# but not everywhere.

tfidf = TfidfVectorizer(
    stop_words="english",
    max_features=30
)

tfidf_matrix = tfidf.fit_transform(df["combined_text"])

tfidf_keywords = pd.DataFrame({
    "keyword": tfidf.get_feature_names_out(),
    "importance_score": tfidf_matrix.sum(axis=0).A1
}).sort_values(by="importance_score", ascending=False)

print("Top TF-IDF keywords in all job posts:")
display(tfidf_keywords.head(30))


# TF-IDF for data-related jobs only
if len(data_jobs) > 0:
    data_tfidf = TfidfVectorizer(
        stop_words="english",
        max_features=30
    )
    
    data_tfidf_matrix = data_tfidf.fit_transform(data_jobs["combined_text"])
    
    data_tfidf_keywords = pd.DataFrame({
        "keyword": data_tfidf.get_feature_names_out(),
        "importance_score": data_tfidf_matrix.sum(axis=0).A1
    }).sort_values(by="importance_score", ascending=False)
    
    print("Top TF-IDF keywords in data-related jobs:")
    display(data_tfidf_keywords.head(30))
else:
    data_tfidf_keywords = pd.DataFrame()
    print("No data-related jobs found for TF-IDF analysis.")


# ================================================================
# 19. VISUALIZATION 1 - JOB COUNT BY CATEGORY
# ================================================================

plt.figure(figsize=(10, 6))
category_count.plot(kind="bar")

plt.title("Number of Job Posts by Category")
plt.xlabel("Job Category")
plt.ylabel("Number of Job Posts")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()

plt.savefig(charts_folder / "job_count_by_category.png", dpi=300)
plt.show()


# ================================================================
# 20. VISUALIZATION 2 - TOP 20 SKILLS
# ================================================================

plt.figure(figsize=(12, 7))
top_skills.sort_values().plot(kind="barh")

plt.title("Top 20 Most Required Skills")
plt.xlabel("Number of Job Posts")
plt.ylabel("Skill")
plt.tight_layout()

plt.savefig(charts_folder / "top_20_skills.png", dpi=300)
plt.show()


# ================================================================
# 21. VISUALIZATION 3 - TOP DATA-RELATED SKILLS
# ================================================================

if len(top_data_skills) > 0:
    plt.figure(figsize=(12, 7))
    top_data_skills.sort_values().plot(kind="barh")

    plt.title("Top 20 Skills for Data-Related Jobs")
    plt.xlabel("Number of Job Posts")
    plt.ylabel("Skill")
    plt.tight_layout()

    plt.savefig(charts_folder / "top_20_data_related_skills.png", dpi=300)
    plt.show()
else:
    print("No data-related skills available for chart.")


# ================================================================
# 22. VISUALIZATION 4 - AVERAGE SKILL COUNT BY CATEGORY
# ================================================================

plt.figure(figsize=(10, 6))
avg_skills_by_category.sort_values().plot(kind="barh")

plt.title("Average Number of Skills Required by Category")
plt.xlabel("Average Skill Count")
plt.ylabel("Category")
plt.tight_layout()

plt.savefig(charts_folder / "average_skill_count_by_category.png", dpi=300)
plt.show()


# ================================================================
# 23. VISUALIZATION 5 - DESCRIPTION WORD COUNT DISTRIBUTION
# ================================================================

plt.figure(figsize=(10, 6))
df["description_word_count"].plot(kind="hist", bins=30)

plt.title("Distribution of Job Description Word Count")
plt.xlabel("Word Count")
plt.ylabel("Number of Job Posts")
plt.tight_layout()

plt.savefig(charts_folder / "description_word_count_distribution.png", dpi=300)
plt.show()


# ================================================================
# 24. SIMPLE STUDENT SKILL GAP ANALYZER
# ================================================================

# Edit this list with your current skills.
# Add the skills you already know.
# Example: Python, SQL, Excel, Power BI, Machine Learning

student_skills = [
    "python",
    "sql",
    "excel",
    "power bi",
    "data analysis",
    "machine learning",
    "statistics"
]

# Convert student skills to lowercase and clean spaces
student_skills = [skill.strip().lower() for skill in student_skills]

print("Student current skills:")
print(student_skills)


# Use top data-related skills if available.
# Otherwise, use top overall skills.
if len(top_data_skills) > 0:
    required_skills = top_data_skills.index.tolist()
else:
    required_skills = top_skills.index.tolist()

matched_skills = sorted(list(set(student_skills).intersection(set(required_skills))))
missing_skills = sorted(list(set(required_skills).difference(set(student_skills))))

match_percentage = round((len(matched_skills) / len(required_skills)) * 100, 2)

print("\nSkill Gap Analysis Result")
print("-------------------------")
print("Total required skills considered:", len(required_skills))
print("Matched skills:", matched_skills)
print("Missing skills:", missing_skills)
print("Skill match percentage:", match_percentage, "%")


# Create a result table
skill_gap_result = pd.DataFrame({
    "required_skill": required_skills,
    "status": ["Matched" if skill in student_skills else "Missing" for skill in required_skills]
})

display(skill_gap_result)


# ================================================================
# 25. RECOMMENDED SKILLS TO LEARN NEXT
# ================================================================

# Recommend the first 10 missing skills based on demand
recommended_skills = missing_skills[:10]

print("Recommended skills to learn next:")
for i, skill in enumerate(recommended_skills, start=1):
    print(i, "-", skill)


# ================================================================
# 26. JOB MATCHING BASED ON STUDENT SKILLS
# ================================================================

# This section finds job posts that best match your current skills.

def skills_to_text(skills):
    """
    Convert a list of skills into one text string.
    """
    return " ".join(skills)

df["skills_text"] = df["skills_list"].apply(skills_to_text)

student_skills_text = " ".join(student_skills)

# Combine student skills with job skills for vectorization
matching_texts = [student_skills_text] + df["skills_text"].tolist()

vectorizer = TfidfVectorizer()
matching_matrix = vectorizer.fit_transform(matching_texts)

# First row is student skills
student_vector = matching_matrix[0]

# Remaining rows are job skills
job_vectors = matching_matrix[1:]

# Calculate similarity between student skills and each job post
similarity_scores = cosine_similarity(student_vector, job_vectors).flatten()

df["student_job_match_score"] = similarity_scores

# Get top 10 matching jobs
top_matching_jobs = df.sort_values(
    by="student_job_match_score",
    ascending=False
).head(10)

print("Top 10 job posts matching your current skills:")
display(top_matching_jobs[[
    "job_id",
    "category",
    "job_title",
    "skill_count",
    "student_job_match_score",
    "skills_list"
]])


# ================================================================
# 27. VISUALIZATION 6 - MATCHED VS MISSING SKILLS
# ================================================================

status_count = skill_gap_result["status"].value_counts()

plt.figure(figsize=(6, 5))
status_count.plot(kind="bar")

plt.title("Matched vs Missing Skills")
plt.xlabel("Skill Status")
plt.ylabel("Number of Skills")
plt.xticks(rotation=0)
plt.tight_layout()

plt.savefig(charts_folder / "matched_vs_missing_skills.png", dpi=300)
plt.show()


# ================================================================
# 28. SAVE CLEANED DATA AND RESULTS
# ================================================================

# Save cleaned full dataset
cleaned_data_path = outputs_folder / "cleaned_all_job_post.csv"
df.to_csv(cleaned_data_path, index=False)

# Save category summary
category_summary_path = outputs_folder / "category_summary.csv"
category_summary.to_csv(category_summary_path)

# Save top skills
top_skills_path = outputs_folder / "top_20_skills.csv"
top_skills.to_csv(top_skills_path, header=["count"])

# Save top data skills
top_data_skills_path = outputs_folder / "top_20_data_related_skills.csv"
top_data_skills.to_csv(top_data_skills_path, header=["count"])

# Save skill gap result
skill_gap_path = outputs_folder / "skill_gap_result.csv"
skill_gap_result.to_csv(skill_gap_path, index=False)

# Save top matching jobs
top_matching_jobs_path = outputs_folder / "top_matching_jobs.csv"
top_matching_jobs.to_csv(top_matching_jobs_path, index=False)

print("Files saved successfully.")
print("Cleaned data saved to:", cleaned_data_path)
print("Category summary saved to:", category_summary_path)
print("Top skills saved to:", top_skills_path)
print("Top data-related skills saved to:", top_data_skills_path)
print("Skill gap result saved to:", skill_gap_path)
print("Top matching jobs saved to:", top_matching_jobs_path)


# ================================================================
# 29. GENERATE SIMPLE TEXT REPORT
# ================================================================

report_text = f"""
InternBridge: Data Science Internship Skill Gap Analyzer

Project Summary
---------------
This project analyzes job postings to identify important skills required in the job market.
The dataset contains {len(df)} job posts and {df['category'].nunique()} job categories.

Dataset Columns
---------------
{', '.join(df.columns)}

Main Findings
-------------
1. Total number of job posts analyzed: {len(df)}
2. Total number of job categories: {df['category'].nunique()}
3. Number of data-related job posts identified: {len(data_jobs)}
4. Percentage of data-related jobs: {round((len(data_jobs) / len(df)) * 100, 2)}%
5. Average number of skills per job post: {round(df['skill_count'].mean(), 2)}
6. Average job description word count: {round(df['description_word_count'].mean(), 2)}

Top 10 Most Required Skills
---------------------------
{top_skills.head(10).to_string()}

Top 10 Data-Related Skills
--------------------------
{top_data_skills.head(10).to_string() if len(top_data_skills) > 0 else 'No data-related skills found.'}

Student Skill Gap Result
------------------------
Student Skills:
{', '.join(student_skills)}

Matched Skills:
{', '.join(matched_skills)}

Missing Skills:
{', '.join(missing_skills)}

Skill Match Percentage:
{match_percentage}%

Recommended Skills to Learn Next
--------------------------------
{', '.join(recommended_skills)}

Conclusion
----------
The analysis helps identify the most demanded skills from job postings.
For a data science undergraduate, this project can be used to understand current job market expectations,
compare personal skills with employer requirements, and plan future learning goals.
"""

report_path = reports_folder / "InternBridge_Summary_Report.txt"

with open(report_path, "w", encoding="utf-8") as file:
    file.write(report_text)

print("Simple text report created successfully.")
print("Report saved to:", report_path)


# ================================================================
# 30. DISPLAY FINAL SUMMARY
# ================================================================

print("================================================")
print("PROJECT COMPLETED SUCCESSFULLY")
print("================================================")
print("Dataset rows:", len(df))
print("Dataset columns:", len(df.columns))
print("Number of categories:", df["category"].nunique())
print("Number of data-related jobs:", len(data_jobs))
print("Skill match percentage:", match_percentage, "%")
print("Charts saved in:", charts_folder)
print("Output CSV files saved in:", outputs_folder)
print("Report saved in:", reports_folder)
print("================================================")
```

```python
# ================================================================
#POWER BI DATA EXPORT
# ================================================================

import pandas as pd
import numpy as np
from pathlib import Path
import ast
import re

from sklearn.feature_extraction.text import TfidfVectorizer, ENGLISH_STOP_WORDS
from sklearn.metrics.pairwise import cosine_similarity


# ------------------------------------------------
# 1. Define folders
# ------------------------------------------------

project_folder = Path(r"D:\Projects\InternBridge")
data_folder = project_folder / "data"
outputs_folder = project_folder / "outputs"
powerbi_folder = project_folder / "powerbi_data"

data_folder.mkdir(exist_ok=True)
outputs_folder.mkdir(exist_ok=True)
powerbi_folder.mkdir(exist_ok=True)

print("Project folder:", project_folder)
print("Power BI folder:", powerbi_folder)


# ------------------------------------------------
# 2. Load dataset
# ------------------------------------------------

cleaned_file = outputs_folder / "cleaned_all_job_post.csv"
raw_csv_file = data_folder / "all_job_post.csv"
raw_excel_file = data_folder / "all_job_post.xlsx"

if cleaned_file.exists():
    df = pd.read_csv(cleaned_file)
    print("Loaded cleaned dataset.")
elif raw_csv_file.exists():
    df = pd.read_csv(raw_csv_file)
    print("Loaded raw CSV dataset.")
elif raw_excel_file.exists():
    df = pd.read_excel(raw_excel_file)
    print("Loaded raw Excel dataset.")
else:
    raise FileNotFoundError("Dataset not found in D:\\Projects\\InternBridge\\data")

print("Dataset shape:", df.shape)


# ------------------------------------------------
# 3. Clean column names
# ------------------------------------------------

df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

required_columns = ["job_id", "category", "job_title", "job_description", "job_skill_set"]

missing_columns = [col for col in required_columns if col not in df.columns]

if missing_columns:
    raise ValueError(f"Missing required columns: {missing_columns}")

df["job_id"] = df["job_id"].astype(str).str.replace(r"\.0$", "", regex=True).str.strip()
df["category"] = df["category"].astype(str).str.strip().str.upper()
df["category_clean"] = df["category"].str.replace("-", " ", regex=False).str.title()
df["job_title"] = df["job_title"].astype(str).str.strip()
df["job_description"] = df["job_description"].astype(str).str.strip()
df["job_skill_set"] = df["job_skill_set"].astype(str).str.strip()

# Create a safer relationship key for Power BI
df["job_key"] = "JOB_" + df["job_id"]

df = df.drop_duplicates()

print("Basic cleaning completed.")


# ------------------------------------------------
# 4. Convert skills into list
# ------------------------------------------------

def convert_skills(skill_text):
    if isinstance(skill_text, list):
        return [str(skill).strip().lower() for skill in skill_text if str(skill).strip() != ""]
    
    try:
        converted = ast.literal_eval(str(skill_text))
        if isinstance(converted, list):
            return [str(skill).strip().lower() for skill in converted if str(skill).strip() != ""]
    except:
        pass
    
    skill_text = str(skill_text).replace("[", "").replace("]", "").replace("'", "").replace('"', "")
    skills = re.split(r",|;|\|", skill_text)
    return [skill.strip().lower() for skill in skills if skill.strip() != ""]

df["skills_list"] = df["job_skill_set"].apply(convert_skills)

df["skills_text"] = df["skills_list"].apply(lambda skills: " ".join(skills))

print("Skills converted.")


# ------------------------------------------------
# 5. Create new columns
# ------------------------------------------------

df["skill_count"] = df["skills_list"].apply(len)

df["description_word_count"] = df["job_description"].apply(
    lambda x: len(str(x).split())
)

df["combined_text"] = (
    df["job_title"].astype(str) + " " +
    df["job_description"].astype(str) + " " +
    df["skills_text"].astype(str)
)


# ------------------------------------------------
# 6. Better data-related job detection
# ------------------------------------------------

# This version avoids marking every job as data-related only because the word "data" appears.
# It focuses on job titles and specific technical/data skills.

title_data_pattern = (
    r"\b(data analyst|data scientist|business intelligence|bi analyst|"
    r"analytics|machine learning|ml engineer|ai engineer|"
    r"database|reporting analyst|dashboard|power bi|tableau|sql|python)\b"
)

skill_data_pattern = (
    r"\b(python|sql|power bi|tableau|machine learning|data analysis|"
    r"statistics|excel|database|reporting|dashboard|analytics|"
    r"business intelligence|data visualization|predictive analytics)\b"
)

description_data_pattern = (
    r"\b(power bi|tableau|machine learning|python|sql|business intelligence|"
    r"data analyst|data scientist|dashboard|data visualization|predictive analytics)\b"
)

df["is_data_related"] = np.where(
    df["job_title"].str.lower().str.contains(title_data_pattern, regex=True, na=False) |
    df["skills_text"].str.lower().str.contains(skill_data_pattern, regex=True, na=False) |
    df["job_description"].str.lower().str.contains(description_data_pattern, regex=True, na=False),
    "Yes",
    "No"
)

print("Data-related job count:")
print(df["is_data_related"].value_counts())


# ------------------------------------------------
# 7. Job matching with your current skills
# ------------------------------------------------

student_skills = [
    "python",
    "sql",
    "excel",
    "power bi",
    "data analysis",
    "machine learning",
    "statistics"
]

student_skills = [skill.strip().lower() for skill in student_skills]

student_skills_text = " ".join(student_skills)

try:
    matching_texts = [student_skills_text] + df["skills_text"].tolist()
    vectorizer = TfidfVectorizer()
    matching_matrix = vectorizer.fit_transform(matching_texts)
    
    student_vector = matching_matrix[0]
    job_vectors = matching_matrix[1:]
    
    similarity_scores = cosine_similarity(student_vector, job_vectors).flatten()
    df["student_job_match_score"] = similarity_scores

except Exception as e:
    print("Job matching failed:", e)
    df["student_job_match_score"] = 0


# ------------------------------------------------
# 8. Main Jobs table
# ------------------------------------------------

jobs_for_powerbi = df[[
    "job_key",
    "job_id",
    "category",
    "category_clean",
    "job_title",
    "skill_count",
    "description_word_count",
    "is_data_related",
    "student_job_match_score"
]].copy()


# ------------------------------------------------
# 9. JobSkills table
# ------------------------------------------------

job_skills_for_powerbi = df[[
    "job_key",
    "job_id",
    "category",
    "category_clean",
    "job_title",
    "skills_list"
]].explode("skills_list")

job_skills_for_powerbi = job_skills_for_powerbi.rename(columns={"skills_list": "skill"})

job_skills_for_powerbi["skill"] = (
    job_skills_for_powerbi["skill"]
    .astype(str)
    .str.strip()
    .str.lower()
)

job_skills_for_powerbi = job_skills_for_powerbi[job_skills_for_powerbi["skill"] != ""]


# ------------------------------------------------
# 10. Category summary
# ------------------------------------------------

category_summary_for_powerbi = df.groupby("category_clean").agg(
    total_jobs=("job_key", "nunique"),
    data_related_jobs=("is_data_related", lambda x: (x == "Yes").sum()),
    average_skill_count=("skill_count", "mean"),
    average_description_words=("description_word_count", "mean"),
    average_match_score=("student_job_match_score", "mean")
).reset_index()

category_summary_for_powerbi["data_related_percentage"] = (
    category_summary_for_powerbi["data_related_jobs"] /
    category_summary_for_powerbi["total_jobs"]
).round(4)


# ------------------------------------------------
# 11. Top skills
# ------------------------------------------------

top_skills_for_powerbi = (
    job_skills_for_powerbi["skill"]
    .value_counts()
    .head(20)
    .reset_index()
)

top_skills_for_powerbi.columns = ["skill", "skill_demand_count"]


# ------------------------------------------------
# 12. Category-wise top skills
# ------------------------------------------------

category_skill_counts = (
    job_skills_for_powerbi
    .groupby(["category_clean", "skill"])
    .size()
    .reset_index(name="skill_demand_count")
)

category_skill_counts["rank_in_category"] = (
    category_skill_counts
    .groupby("category_clean")["skill_demand_count"]
    .rank(method="first", ascending=False)
)

category_top_skills_for_powerbi = category_skill_counts[
    category_skill_counts["rank_in_category"] <= 10
].sort_values(["category_clean", "rank_in_category"])


# ------------------------------------------------
# 13. Data-related jobs and skills
# ------------------------------------------------

data_jobs = df[df["is_data_related"] == "Yes"].copy()

data_related_jobs_for_powerbi = data_jobs[[
    "job_key",
    "job_id",
    "category_clean",
    "job_title",
    "skill_count",
    "description_word_count",
    "student_job_match_score"
]].copy()

data_skills_for_powerbi = data_jobs[[
    "job_key",
    "job_id",
    "category_clean",
    "job_title",
    "skills_list"
]].explode("skills_list")

data_skills_for_powerbi = data_skills_for_powerbi.rename(columns={"skills_list": "skill"})

data_skills_for_powerbi["skill"] = (
    data_skills_for_powerbi["skill"]
    .astype(str)
    .str.strip()
    .str.lower()
)

data_skills_for_powerbi = data_skills_for_powerbi[data_skills_for_powerbi["skill"] != ""]

top_data_skills_for_powerbi = (
    data_skills_for_powerbi["skill"]
    .value_counts()
    .head(20)
    .reset_index()
)

top_data_skills_for_powerbi.columns = ["skill", "skill_demand_count"]


# ------------------------------------------------
# 14. Skill gap analysis
# ------------------------------------------------

required_skills = top_data_skills_for_powerbi["skill"].tolist()

matched_skills = sorted(list(set(student_skills).intersection(set(required_skills))))
missing_skills = sorted(list(set(required_skills).difference(set(student_skills))))

match_percentage = round((len(matched_skills) / len(required_skills)) * 100, 2) if required_skills else 0

skill_gap_for_powerbi = pd.DataFrame({
    "required_skill": required_skills,
    "status": ["Matched" if skill in student_skills else "Missing" for skill in required_skills]
})

skill_gap_for_powerbi["skill_rank"] = range(1, len(skill_gap_for_powerbi) + 1)

skill_gap_summary_for_powerbi = pd.DataFrame({
    "metric": ["Matched Skills", "Missing Skills", "Skill Match Percentage"],
    "value": [
        (skill_gap_for_powerbi["status"] == "Matched").sum(),
        (skill_gap_for_powerbi["status"] == "Missing").sum(),
        match_percentage
    ]
})


# ------------------------------------------------
# 15. Top matching jobs
# ------------------------------------------------

top_matching_jobs_for_powerbi = df.sort_values(
    by="student_job_match_score",
    ascending=False
).head(15)[[
    "job_key",
    "job_id",
    "category_clean",
    "job_title",
    "skill_count",
    "student_job_match_score",
    "is_data_related"
]].copy()


# ------------------------------------------------
# 16. Better TF-IDF keyword analysis
# ------------------------------------------------

custom_stop_words = set(ENGLISH_STOP_WORDS).union({
    "sales", "business", "management", "manager", "experience",
    "work", "team", "company", "customer", "support", "job",
    "role", "employee", "hr", "financial", "development",
    "service", "including", "ability", "skills", "required",
    "responsibilities", "requirements", "year", "years"
})

try:
    tfidf = TfidfVectorizer(
        stop_words=list(custom_stop_words),
        max_features=30,
        ngram_range=(1, 2)
    )
    
    tfidf_matrix = tfidf.fit_transform(df["combined_text"])
    
    tfidf_keywords_for_powerbi = pd.DataFrame({
        "keyword": tfidf.get_feature_names_out(),
        "importance_score": tfidf_matrix.sum(axis=0).A1
    }).sort_values("importance_score", ascending=False)

except Exception as e:
    print("TF-IDF all jobs failed:", e)
    tfidf_keywords_for_powerbi = pd.DataFrame(columns=["keyword", "importance_score"])

try:
    if len(data_jobs) > 0:
        data_tfidf = TfidfVectorizer(
            stop_words=list(custom_stop_words),
            max_features=30,
            ngram_range=(1, 2)
        )
        
        data_tfidf_matrix = data_tfidf.fit_transform(data_jobs["combined_text"])
        
        data_tfidf_keywords_for_powerbi = pd.DataFrame({
            "keyword": data_tfidf.get_feature_names_out(),
            "importance_score": data_tfidf_matrix.sum(axis=0).A1
        }).sort_values("importance_score", ascending=False)
    else:
        data_tfidf_keywords_for_powerbi = pd.DataFrame(columns=["keyword", "importance_score"])

except Exception as e:
    print("TF-IDF data jobs failed:", e)
    data_tfidf_keywords_for_powerbi = pd.DataFrame(columns=["keyword", "importance_score"])


# ------------------------------------------------
# 17. KPI summary
# ------------------------------------------------

kpi_summary_for_powerbi = pd.DataFrame({
    "metric": [
        "Total Job Posts",
        "Total Categories",
        "Data Related Jobs",
        "Data Related Percentage",
        "Average Skill Count",
        "Average Description Words",
        "Skill Match Percentage"
    ],
    "value": [
        df["job_key"].nunique(),
        df["category_clean"].nunique(),
        (df["is_data_related"] == "Yes").sum(),
        round(((df["is_data_related"] == "Yes").sum() / df["job_key"].nunique()) * 100, 2),
        round(df["skill_count"].mean(), 2),
        round(df["description_word_count"].mean(), 2),
        match_percentage
    ]
})


# ------------------------------------------------
# 18. Export all files
# ------------------------------------------------

jobs_for_powerbi.to_csv(powerbi_folder / "jobs_for_powerbi.csv", index=False)
job_skills_for_powerbi.to_csv(powerbi_folder / "job_skills_for_powerbi.csv", index=False)
category_summary_for_powerbi.to_csv(powerbi_folder / "category_summary_for_powerbi.csv", index=False)
top_skills_for_powerbi.to_csv(powerbi_folder / "top_skills_for_powerbi.csv", index=False)
category_top_skills_for_powerbi.to_csv(powerbi_folder / "category_top_skills_for_powerbi.csv", index=False)
data_related_jobs_for_powerbi.to_csv(powerbi_folder / "data_related_jobs_for_powerbi.csv", index=False)
top_data_skills_for_powerbi.to_csv(powerbi_folder / "top_data_skills_for_powerbi.csv", index=False)
skill_gap_for_powerbi.to_csv(powerbi_folder / "skill_gap_for_powerbi.csv", index=False)
skill_gap_summary_for_powerbi.to_csv(powerbi_folder / "skill_gap_summary_for_powerbi.csv", index=False)
top_matching_jobs_for_powerbi.to_csv(powerbi_folder / "top_matching_jobs_for_powerbi.csv", index=False)
tfidf_keywords_for_powerbi.to_csv(powerbi_folder / "tfidf_keywords_for_powerbi.csv", index=False)
data_tfidf_keywords_for_powerbi.to_csv(powerbi_folder / "data_tfidf_keywords_for_powerbi.csv", index=False)
kpi_summary_for_powerbi.to_csv(powerbi_folder / "kpi_summary_for_powerbi.csv", index=False)

df.to_csv(outputs_folder / "cleaned_all_job_post.csv", index=False)

print("================================================")
print("IMPROVED POWER BI DATA EXPORT COMPLETED")
print("Files saved to:", powerbi_folder)
print("Data-related jobs:", (df["is_data_related"] == "Yes").sum())
print("Skill match percentage:", match_percentage, "%")
print("================================================")

for file in powerbi_folder.glob("*.csv"):
    print("-", file.name)
```

```python
# ================================================================
# FINAL POWER BI DATA FIX
# Project: InternBridge
# Purpose:
# 1. Recreate clean CSV files with correct headers
# 2. Create one Excel workbook with separate sheets for Power BI
# 3. Avoid Power BI showing Column1, Column2, Column3...
# ================================================================

import pandas as pd
import numpy as np
from pathlib import Path
import ast
import re

from sklearn.feature_extraction.text import TfidfVectorizer, ENGLISH_STOP_WORDS
from sklearn.metrics.pairwise import cosine_similarity


# ------------------------------------------------
# 1. Define project folders
# ------------------------------------------------

project_folder = Path(r"D:\Projects\InternBridge")
data_folder = project_folder / "data"
outputs_folder = project_folder / "outputs"

# New clean folder for final Power BI files
powerbi_final_folder = project_folder / "powerbi_final_data"
powerbi_final_folder.mkdir(exist_ok=True)

print("Project folder:", project_folder)
print("Final Power BI folder:", powerbi_final_folder)


# ------------------------------------------------
# 2. Load original dataset
# ------------------------------------------------

raw_csv_file = data_folder / "all_job_post.csv"
raw_excel_file = data_folder / "all_job_post.xlsx"
cleaned_file = outputs_folder / "cleaned_all_job_post.csv"

# Prefer original data because it is cleaner for rebuilding everything
if raw_csv_file.exists():
    df = pd.read_csv(raw_csv_file)
    print("Loaded original CSV dataset.")
elif raw_excel_file.exists():
    df = pd.read_excel(raw_excel_file)
    print("Loaded original Excel dataset.")
elif cleaned_file.exists():
    df = pd.read_csv(cleaned_file)
    print("Loaded cleaned dataset.")
else:
    raise FileNotFoundError("Dataset not found. Please check D:\\Projects\\InternBridge\\data")

print("Dataset shape:", df.shape)


# ------------------------------------------------
# 3. Clean column names
# ------------------------------------------------

df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

required_columns = [
    "job_id",
    "category",
    "job_title",
    "job_description",
    "job_skill_set"
]

missing_columns = [col for col in required_columns if col not in df.columns]

if missing_columns:
    raise ValueError(f"Missing required columns: {missing_columns}")

print("Columns checked successfully.")


# ------------------------------------------------
# 4. Basic cleaning
# ------------------------------------------------

df["job_id"] = df["job_id"].astype(str).str.replace(r"\.0$", "", regex=True).str.strip()
df["category"] = df["category"].astype(str).str.strip().str.upper()
df["category_clean"] = df["category"].str.replace("-", " ", regex=False).str.title()
df["job_title"] = df["job_title"].astype(str).str.strip()
df["job_description"] = df["job_description"].astype(str).str.strip()
df["job_skill_set"] = df["job_skill_set"].astype(str).str.strip()

# Create clean relationship key
df["job_key"] = "JOB_" + df["job_id"]

df = df.drop_duplicates()

print("Basic cleaning completed.")


# ------------------------------------------------
# 5. Convert skill text into Python lists
# ------------------------------------------------

def convert_skills(skill_text):
    if isinstance(skill_text, list):
        return [str(skill).strip().lower() for skill in skill_text if str(skill).strip() != ""]
    
    try:
        converted = ast.literal_eval(str(skill_text))
        if isinstance(converted, list):
            return [str(skill).strip().lower() for skill in converted if str(skill).strip() != ""]
    except:
        pass
    
    skill_text = str(skill_text).replace("[", "").replace("]", "").replace("'", "").replace('"', "")
    skills = re.split(r",|;|\|", skill_text)
    return [skill.strip().lower() for skill in skills if skill.strip() != ""]


df["skills_list"] = df["job_skill_set"].apply(convert_skills)
df["skills_text"] = df["skills_list"].apply(lambda skills: " ".join(skills))

print("Skills converted successfully.")


# ------------------------------------------------
# 6. Create useful columns
# ------------------------------------------------

df["skill_count"] = df["skills_list"].apply(len)

df["description_word_count"] = df["job_description"].apply(
    lambda x: len(str(x).split())
)

df["combined_text"] = (
    df["job_title"].astype(str) + " " +
    df["job_description"].astype(str) + " " +
    df["skills_text"].astype(str)
)


# ------------------------------------------------
# 7. Data-related job detection
# ------------------------------------------------

title_data_pattern = (
    r"\b(?:data analyst|data scientist|business intelligence|bi analyst|"
    r"analytics|machine learning|ml engineer|ai engineer|"
    r"database|reporting analyst|dashboard|power bi|tableau|sql|python)\b"
)

skill_data_pattern = (
    r"\b(?:python|sql|power bi|tableau|machine learning|data analysis|"
    r"statistics|excel|database|reporting|dashboard|analytics|"
    r"business intelligence|data visualization|predictive analytics)\b"
)

description_data_pattern = (
    r"\b(?:power bi|tableau|machine learning|python|sql|business intelligence|"
    r"data analyst|data scientist|dashboard|data visualization|predictive analytics)\b"
)

df["is_data_related"] = np.where(
    df["job_title"].str.lower().str.contains(title_data_pattern, regex=True, na=False) |
    df["skills_text"].str.lower().str.contains(skill_data_pattern, regex=True, na=False) |
    df["job_description"].str.lower().str.contains(description_data_pattern, regex=True, na=False),
    "Yes",
    "No"
)

print("Data-related job count:")
print(df["is_data_related"].value_counts())


# ------------------------------------------------
# 8. Student skills and job matching
# ------------------------------------------------

student_skills = [
    "python",
    "sql",
    "excel",
    "power bi",
    "data analysis",
    "machine learning",
    "statistics"
]

student_skills = [skill.strip().lower() for skill in student_skills]
student_skills_text = " ".join(student_skills)

try:
    matching_texts = [student_skills_text] + df["skills_text"].tolist()
    vectorizer = TfidfVectorizer()
    matching_matrix = vectorizer.fit_transform(matching_texts)

    student_vector = matching_matrix[0]
    job_vectors = matching_matrix[1:]

    similarity_scores = cosine_similarity(student_vector, job_vectors).flatten()
    df["student_job_match_score"] = similarity_scores

except Exception as e:
    print("Job matching failed:", e)
    df["student_job_match_score"] = 0


# ------------------------------------------------
# 9. Create Jobs table
# ------------------------------------------------

jobs = df[[
    "job_key",
    "job_id",
    "category",
    "category_clean",
    "job_title",
    "skill_count",
    "description_word_count",
    "is_data_related",
    "student_job_match_score"
]].copy()


# ------------------------------------------------
# 10. Create JobSkills table
# ------------------------------------------------

job_skills = df[[
    "job_key",
    "job_id",
    "category",
    "category_clean",
    "job_title",
    "skills_list"
]].explode("skills_list")

job_skills = job_skills.rename(columns={"skills_list": "skill"})

job_skills["skill"] = (
    job_skills["skill"]
    .astype(str)
    .str.strip()
    .str.lower()
)

job_skills = job_skills[job_skills["skill"] != ""]


# ------------------------------------------------
# 11. Create CategorySummary table
# ------------------------------------------------

category_summary = df.groupby("category_clean").agg(
    total_jobs=("job_key", "nunique"),
    data_related_jobs=("is_data_related", lambda x: (x == "Yes").sum()),
    average_skill_count=("skill_count", "mean"),
    average_description_words=("description_word_count", "mean"),
    average_match_score=("student_job_match_score", "mean")
).reset_index()

category_summary["data_related_percentage"] = (
    category_summary["data_related_jobs"] /
    category_summary["total_jobs"] * 100
).round(2)

# Add compatibility columns
category_summary["category"] = category_summary["category_clean"]
category_summary["number_of_jobs"] = category_summary["total_jobs"]


# ------------------------------------------------
# 12. Create TopSkills table
# ------------------------------------------------

top_skills = (
    job_skills["skill"]
    .value_counts()
    .head(20)
    .reset_index()
)

top_skills.columns = ["skill", "skill_demand_count"]

# Add compatibility column
top_skills["count"] = top_skills["skill_demand_count"]


# ------------------------------------------------
# 13. Create CategoryTopSkills table
# ------------------------------------------------

category_skill_counts = (
    job_skills
    .groupby(["category_clean", "skill"])
    .size()
    .reset_index(name="skill_demand_count")
)

category_skill_counts["rank_in_category"] = (
    category_skill_counts
    .groupby("category_clean")["skill_demand_count"]
    .rank(method="first", ascending=False)
)

category_top_skills = category_skill_counts[
    category_skill_counts["rank_in_category"] <= 10
].sort_values(["category_clean", "rank_in_category"])


# ------------------------------------------------
# 14. Create data-related tables
# ------------------------------------------------

data_jobs = df[df["is_data_related"] == "Yes"].copy()

data_related_jobs = data_jobs[[
    "job_key",
    "job_id",
    "category",
    "category_clean",
    "job_title",
    "skill_count",
    "description_word_count",
    "student_job_match_score",
    "is_data_related"
]].copy()

data_skills = data_jobs[[
    "job_key",
    "job_id",
    "category",
    "category_clean",
    "job_title",
    "skills_list"
]].explode("skills_list")

data_skills = data_skills.rename(columns={"skills_list": "skill"})

data_skills["skill"] = (
    data_skills["skill"]
    .astype(str)
    .str.strip()
    .str.lower()
)

data_skills = data_skills[data_skills["skill"] != ""]

top_data_skills = (
    data_skills["skill"]
    .value_counts()
    .head(20)
    .reset_index()
)

top_data_skills.columns = ["skill", "skill_demand_count"]

# Add compatibility column
top_data_skills["count"] = top_data_skills["skill_demand_count"]


# ------------------------------------------------
# 15. Create SkillGap table
# ------------------------------------------------

required_skills = top_data_skills["skill"].tolist()

matched_skills = sorted(list(set(student_skills).intersection(set(required_skills))))
missing_skills = sorted(list(set(required_skills).difference(set(student_skills))))

match_percentage = round((len(matched_skills) / len(required_skills)) * 100, 2) if required_skills else 0

skill_gap = pd.DataFrame({
    "required_skill": required_skills,
    "status": ["Matched" if skill in student_skills else "Missing" for skill in required_skills]
})

skill_gap["skill_rank"] = range(1, len(skill_gap) + 1)

skill_gap_summary = pd.DataFrame({
    "metric": [
        "Matched Skills",
        "Missing Skills",
        "Skill Match Percentage"
    ],
    "value": [
        (skill_gap["status"] == "Matched").sum(),
        (skill_gap["status"] == "Missing").sum(),
        match_percentage
    ]
})


# ------------------------------------------------
# 16. Create TopMatchingJobs table
# ------------------------------------------------

top_matching_jobs = df.sort_values(
    by="student_job_match_score",
    ascending=False
).head(15)[[
    "job_key",
    "job_id",
    "category",
    "category_clean",
    "job_title",
    "skill_count",
    "student_job_match_score",
    "is_data_related"
]].copy()


# ------------------------------------------------
# 17. Create TF-IDF keyword tables
# ------------------------------------------------

custom_stop_words = set(ENGLISH_STOP_WORDS).union({
    "sales", "business", "management", "manager", "experience",
    "work", "team", "company", "customer", "support", "job",
    "role", "employee", "hr", "financial", "development",
    "service", "including", "ability", "skills", "required",
    "responsibilities", "requirements", "year", "years"
})

try:
    tfidf = TfidfVectorizer(
        stop_words=list(custom_stop_words),
        max_features=30,
        ngram_range=(1, 2)
    )

    tfidf_matrix = tfidf.fit_transform(df["combined_text"])

    tfidf_keywords = pd.DataFrame({
        "keyword": tfidf.get_feature_names_out(),
        "importance_score": tfidf_matrix.sum(axis=0).A1
    }).sort_values("importance_score", ascending=False)

except Exception as e:
    print("TF-IDF all jobs failed:", e)
    tfidf_keywords = pd.DataFrame(columns=["keyword", "importance_score"])


try:
    if len(data_jobs) > 0:
        data_tfidf = TfidfVectorizer(
            stop_words=list(custom_stop_words),
            max_features=30,
            ngram_range=(1, 2)
        )

        data_tfidf_matrix = data_tfidf.fit_transform(data_jobs["combined_text"])

        data_tfidf_keywords = pd.DataFrame({
            "keyword": data_tfidf.get_feature_names_out(),
            "importance_score": data_tfidf_matrix.sum(axis=0).A1
        }).sort_values("importance_score", ascending=False)
    else:
        data_tfidf_keywords = pd.DataFrame(columns=["keyword", "importance_score"])

except Exception as e:
    print("TF-IDF data jobs failed:", e)
    data_tfidf_keywords = pd.DataFrame(columns=["keyword", "importance_score"])


# ------------------------------------------------
# 18. Create KPI Summary table
# ------------------------------------------------

kpi_summary = pd.DataFrame({
    "metric": [
        "Total Job Posts",
        "Total Categories",
        "Data Related Jobs",
        "Data Related Percentage",
        "Average Skill Count",
        "Average Description Words",
        "Skill Match Percentage"
    ],
    "value": [
        df["job_key"].nunique(),
        df["category_clean"].nunique(),
        (df["is_data_related"] == "Yes").sum(),
        round(((df["is_data_related"] == "Yes").sum() / df["job_key"].nunique()) * 100, 2),
        round(df["skill_count"].mean(), 2),
        round(df["description_word_count"].mean(), 2),
        match_percentage
    ]
})


# ------------------------------------------------
# 19. Save all clean CSV files with proper headers
# ------------------------------------------------

tables = {
    "Jobs": jobs,
    "JobSkills": job_skills,
    "CategorySummary": category_summary,
    "TopSkills": top_skills,
    "CategoryTopSkills": category_top_skills,
    "DataRelatedJobs": data_related_jobs,
    "TopDataSkills": top_data_skills,
    "SkillGap": skill_gap,
    "SkillGapSummary": skill_gap_summary,
    "TopMatchingJobs": top_matching_jobs,
    "TFIDFKeywords": tfidf_keywords,
    "DataTFIDFKeywords": data_tfidf_keywords,
    "KPISummary": kpi_summary
}

for table_name, table_df in tables.items():
    csv_path = powerbi_final_folder / f"{table_name}.csv"
    table_df.to_csv(csv_path, index=False, encoding="utf-8-sig")
    print("Saved CSV:", csv_path.name)


# ------------------------------------------------
# 20. Save one Excel workbook for easier Power BI import
# ------------------------------------------------

excel_path = powerbi_final_folder / "InternBridge_PowerBI_Final_Data.xlsx"

with pd.ExcelWriter(excel_path, engine="openpyxl") as writer:
    for sheet_name, table_df in tables.items():
        # Excel sheet names must be 31 characters or less
        safe_sheet_name = sheet_name[:31]
        table_df.to_excel(writer, sheet_name=safe_sheet_name, index=False)

print("Saved Excel workbook:", excel_path)


# ------------------------------------------------
# 21. Save cleaned full dataset also
# ------------------------------------------------

df.to_csv(outputs_folder / "cleaned_all_job_post.csv", index=False, encoding="utf-8-sig")


# ------------------------------------------------
# 22. Final confirmation
# ------------------------------------------------

print("================================================")
print("FINAL POWER BI DATA FIX COMPLETED SUCCESSFULLY")
print("Clean files saved in:", powerbi_final_folder)
print("Excel file:", excel_path)
print("Total jobs:", df["job_key"].nunique())
print("Data-related jobs:", (df["is_data_related"] == "Yes").sum())
print("Skill match percentage:", match_percentage, "%")
print("================================================")
```

```python
# ================================================================
# InternBridge: Data Science Internship Skill Gap Analyzer
# Improved Data Science Role Filtering + Power BI Export
# Author: Chamodi Kumarage
# Purpose:
#   Fix the issue where all data-related jobs are treated as data science jobs.
#   This script separates true data science / analytics roles from other jobs.
# ================================================================

import ast
import re
from pathlib import Path

import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS, TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


# ------------------------------------------------
# 1. Set project folders
# ------------------------------------------------
# Run this file from D:\Projects\InternBridge.
# If you run it from the notebooks folder, it will still find the project folder.
current_folder = Path.cwd()
if current_folder.name.lower() in ["notebooks", "scripts"]:
    project_folder = current_folder.parent
else:
    project_folder = current_folder

# For a fixed Windows path, uncomment the next line:
# project_folder = Path(r"D:\Projects\InternBridge")

data_folder = project_folder / "data"
outputs_folder = project_folder / "outputs"
reports_folder = project_folder / "reports"
powerbi_folder = project_folder / "powerbi_ds_final_data"

outputs_folder.mkdir(exist_ok=True)
reports_folder.mkdir(exist_ok=True)
powerbi_folder.mkdir(exist_ok=True)

print("Project folder:", project_folder)
print("Power BI output folder:", powerbi_folder)


# ------------------------------------------------
# 2. Load original data
# ------------------------------------------------
raw_csv_file = data_folder / "all_job_post.csv"
raw_excel_file = data_folder / "all_job_post.xlsx"

if raw_csv_file.exists():
    df = pd.read_csv(raw_csv_file)
    print("Loaded original CSV dataset.")
elif raw_excel_file.exists():
    df = pd.read_excel(raw_excel_file)
    print("Loaded original Excel dataset.")
else:
    raise FileNotFoundError("Please place all_job_post.csv or all_job_post.xlsx inside the data folder.")

print("Original dataset shape:", df.shape)


# ------------------------------------------------
# 3. Clean column names and required columns
# ------------------------------------------------
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

required_columns = [
    "job_id",
    "category",
    "job_title",
    "job_description",
    "job_skill_set"
]

missing_columns = [col for col in required_columns if col not in df.columns]
if missing_columns:
    raise ValueError(f"Missing required columns: {missing_columns}")

# Basic text cleaning
for col in ["job_id", "category", "job_title", "job_description", "job_skill_set"]:
    df[col] = df[col].fillna("").astype(str).str.strip()

df["job_id"] = df["job_id"].str.replace(r"\.0$", "", regex=True)
df["category"] = df["category"].str.upper()
df["category_clean"] = df["category"].str.replace("-", " ", regex=False).str.title()
df["job_key"] = "JOB_" + df["job_id"]

df = df.drop_duplicates(subset=["job_id", "job_title", "job_description", "job_skill_set"]).copy()

print("Dataset shape after cleaning:", df.shape)


# ------------------------------------------------
# 4. Convert skill text into clean skill lists
# ------------------------------------------------
SKILL_NORMALIZATION = {
    "microsoft excel": "excel",
    "ms excel": "excel",
    "advanced excel": "excel",
    "microsoft power bi": "power bi",
    "powerbi": "power bi",
    "ms power bi": "power bi",
    "microsoft sql server": "sql",
    "sql server": "sql",
    "structured query language": "sql",
    "python programming": "python",
    "r programming": "r",
    "statistical analysis": "statistics",
    "data visualisation": "data visualization",
    "data visualizations": "data visualization",
    "ml": "machine learning",
    "ai": "artificial intelligence",
}


def normalize_skill(skill):
    skill = str(skill).strip().lower()
    skill = re.sub(r"\s+", " ", skill)
    return SKILL_NORMALIZATION.get(skill, skill)


def convert_skills(skill_text):
    if isinstance(skill_text, list):
        raw_skills = skill_text
    else:
        try:
            converted = ast.literal_eval(str(skill_text))
            raw_skills = converted if isinstance(converted, list) else []
        except Exception:
            cleaned_text = str(skill_text).replace("[", "").replace("]", "")
            cleaned_text = cleaned_text.replace("'", "").replace('"', "")
            raw_skills = re.split(r",|;|\|", cleaned_text)

    cleaned_skills = []
    for skill in raw_skills:
        normalized = normalize_skill(skill)
        if normalized and normalized not in cleaned_skills:
            cleaned_skills.append(normalized)
    return cleaned_skills


df["skills_list"] = df["job_skill_set"].apply(convert_skills)
df["skills_text"] = df["skills_list"].apply(lambda skills: " ".join(skills))
df["skill_count"] = df["skills_list"].apply(len)
df["description_word_count"] = df["job_description"].apply(lambda x: len(str(x).split()))
df["combined_text"] = (
    df["job_title"].astype(str) + " " +
    df["job_description"].astype(str) + " " +
    df["skills_text"].astype(str)
)


# ------------------------------------------------
# 5. Correct role classification
# ------------------------------------------------
# Important idea:
#   A job is NOT a data science role only because it has Excel, SQL, reports, or data entry.
#   We first check strong role titles.
#   Ambiguous jobs are marked as Review, not automatically used in the analysis.

ROLE_TITLE_PATTERNS = [
    ("Data Science", r"\b(data scientist|data science|decision scientist|applied scientist|predictive model(?:ing)?|statistical model(?:ing)?)\b"),
    ("Machine Learning / AI", r"\b(machine learning|ml engineer|ai engineer|artificial intelligence|deep learning|nlp|computer vision)\b"),
    ("Data Engineering", r"\b(data engineer|analytics engineer|etl developer|data warehouse|data warehousing|data pipeline|big data engineer|data platform engineer)\b"),
    ("Data Analytics", r"\b(data analyst|analytics analyst|data analytics|data & analytics|data solutions analyst|data visualization analyst|statistical analyst|quantitative analyst)\b"),
    ("Business Intelligence", r"\b(business intelligence|bi analyst|power bi developer|tableau developer|dashboard developer|reporting analyst)\b"),
]

EXCLUDE_TITLE_PATTERN = (
    r"\b(human resources|hr|sales|accounting|auditor|audit|it support|help desk|"
    r"security analyst|cyber|network|system administrator|product owner|program manager|"
    r"project manager|software engineer|developer|programmer|operations analyst|"
    r"financial analyst|finance analyst|fp&a analyst|business development)\b"
)

ADVANCED_DATA_SKILLS = [
    "python", "r", "sql", "machine learning", "artificial intelligence",
    "statistics", "predictive modeling", "data modeling", "data warehousing",
    "etl", "elt", "data pipeline", "data visualization", "power bi", "tableau",
    "spark", "databricks", "big data", "azure", "aws", "data governance"
]


def count_advanced_data_skills(skills_text):
    score = 0
    for skill in ADVANCED_DATA_SKILLS:
        if re.search(r"\b" + re.escape(skill) + r"\b", str(skills_text).lower()):
            score += 1
    return score


def classify_role(row):
    title = str(row["job_title"]).lower()
    description = str(row["job_description"]).lower()
    skills_text = str(row["skills_text"]).lower()

    # Strong title evidence: automatically accepted as a target data science/analytics role.
    for family, pattern in ROLE_TITLE_PATTERNS:
        if re.search(pattern, title, flags=re.IGNORECASE):
            return pd.Series(["Yes", family, "Strong title match"])

    skill_score = count_advanced_data_skills(skills_text)
    has_data_context = bool(re.search(
        r"\b(data|analytics|business intelligence|machine learning|predictive|statistics|dashboard|reporting)\b",
        title + " " + description,
        flags=re.IGNORECASE
    ))
    has_weak_role_word = bool(re.search(r"\b(analyst|manager|specialist|consultant|engineer)\b", title, flags=re.IGNORECASE))
    has_excluded_title = bool(re.search(EXCLUDE_TITLE_PATTERN, title, flags=re.IGNORECASE))
    has_data_word_in_title = bool(re.search(r"\b(data|analytics|business intelligence|machine learning|bi)\b", title, flags=re.IGNORECASE))

    # Ambiguous high-data jobs should be reviewed manually.
    # They are not included in final skill-gap analysis unless you change Review to Yes.
    if has_data_context and has_weak_role_word and skill_score >= 5:
        if has_excluded_title and not has_data_word_in_title:
            return pd.Series(["Review", "Possible Data Role", f"Ambiguous: advanced data skill score = {skill_score}"])
        return pd.Series(["Review", "Possible Data Role", f"Needs manual review: advanced data skill score = {skill_score}"])

    return pd.Series(["No", "Not Data Science", f"Not enough evidence: advanced data skill score = {skill_score}"])


df[["is_true_data_science_role", "data_role_family", "classification_reason"]] = df.apply(classify_role, axis=1)


# ------------------------------------------------
# 6. Internship / seniority detection
# ------------------------------------------------
def classify_seniority(title):
    title = str(title).lower()
    if re.search(r"\b(intern|internship|trainee)\b", title):
        return "Internship"
    if re.search(r"\b(entry level|entry-level|junior|graduate|associate)\b", title):
        return "Entry / Junior"
    if re.search(r"\b(senior|sr\.?|lead|principal|manager|director|vp|vice president|chief|head)\b", title):
        return "Senior / Management"
    return "Mid / Not Specified"


df["seniority_level"] = df["job_title"].apply(classify_seniority)
df["is_internship_role"] = np.where(df["seniority_level"] == "Internship", "Yes", "No")

# Main filter for the project analysis
analysis_jobs = df[df["is_true_data_science_role"] == "Yes"].copy()
review_jobs = df[df["is_true_data_science_role"] == "Review"].copy()
true_ds_internships = analysis_jobs[analysis_jobs["is_internship_role"] == "Yes"].copy()

print("Role classification result:")
print(df["is_true_data_science_role"].value_counts())
print("True data science / analytics roles:", len(analysis_jobs))
print("True data science / analytics internships:", len(true_ds_internships))
print("Manual review roles:", len(review_jobs))


# ------------------------------------------------
# 7. Skill demand analysis using ONLY true data science/analytics roles
# ------------------------------------------------
def explode_skills(input_df):
    if input_df.empty:
        return pd.DataFrame(columns=["job_key", "job_id", "category", "category_clean", "job_title", "data_role_family", "skill"])
    exploded = input_df[[
        "job_key", "job_id", "category", "category_clean", "job_title", "data_role_family", "skills_list"
    ]].explode("skills_list")
    exploded = exploded.rename(columns={"skills_list": "skill"})
    exploded["skill"] = exploded["skill"].astype(str).str.strip().str.lower()
    exploded = exploded[exploded["skill"] != ""]
    return exploded


all_job_skills = explode_skills(df)
ds_job_skills = explode_skills(analysis_jobs)

top_ds_skills = (
    ds_job_skills["skill"].value_counts().head(20).reset_index()
    if not ds_job_skills.empty else pd.DataFrame(columns=["skill", "skill_demand_count"])
)
if not top_ds_skills.empty:
    top_ds_skills.columns = ["skill", "skill_demand_count"]
    top_ds_skills["count"] = top_ds_skills["skill_demand_count"]


# ------------------------------------------------
# 8. Student skill gap analysis
# ------------------------------------------------
# Edit this list with your current skills.
student_skills = [
    "python",
    "sql",
    "excel",
    "power bi",
    "data analysis",
    "machine learning",
    "statistics"
]
student_skills = [normalize_skill(skill) for skill in student_skills]
student_skill_set = set(student_skills)

required_skills = top_ds_skills["skill"].tolist() if not top_ds_skills.empty else []
matched_skills = sorted(set(required_skills).intersection(student_skill_set))
missing_skills = sorted(set(required_skills).difference(student_skill_set))
match_percentage = round((len(matched_skills) / len(required_skills)) * 100, 2) if required_skills else 0

skill_gap = pd.DataFrame({
    "required_skill": required_skills,
    "status": ["Matched" if skill in student_skill_set else "Missing" for skill in required_skills]
})
skill_gap["skill_rank"] = range(1, len(skill_gap) + 1)

recommendations = skill_gap[skill_gap["status"] == "Missing"].copy()
if not recommendations.empty:
    recommendations = recommendations.merge(top_ds_skills[["skill", "skill_demand_count"]], left_on="required_skill", right_on="skill", how="left")
    recommendations = recommendations.drop(columns=["skill"])
    recommendations["priority_score"] = recommendations["skill_demand_count"].rank(method="dense", ascending=False).astype(int)
    recommendations = recommendations.sort_values(["priority_score", "skill_rank"])

skill_gap_summary = pd.DataFrame({
    "metric": ["Required Skills Considered", "Matched Skills", "Missing Skills", "Skill Match Percentage"],
    "value": [len(required_skills), len(matched_skills), len(missing_skills), match_percentage]
})


# ------------------------------------------------
# 9. Job matching using ONLY true data science/analytics roles
# ------------------------------------------------
def calculate_job_match(row):
    job_skill_set = set(row["skills_list"])
    matched = sorted(job_skill_set.intersection(student_skill_set))
    missing = sorted(job_skill_set.difference(student_skill_set))
    overlap_score = (len(matched) / len(job_skill_set) * 100) if job_skill_set else 0
    return pd.Series([len(matched), ", ".join(matched), ", ".join(missing), round(overlap_score, 2)])

if not analysis_jobs.empty:
    analysis_jobs[["matched_skill_count", "matched_skills", "missing_skills_for_job", "skill_overlap_percentage"]] = analysis_jobs.apply(calculate_job_match, axis=1)

    try:
        matching_texts = [" ".join(student_skills)] + analysis_jobs["skills_text"].tolist()
        vectorizer = TfidfVectorizer()
        matching_matrix = vectorizer.fit_transform(matching_texts)
        similarity_scores = cosine_similarity(matching_matrix[0], matching_matrix[1:]).flatten()
        analysis_jobs["tfidf_similarity"] = similarity_scores
    except Exception:
        analysis_jobs["tfidf_similarity"] = 0

    analysis_jobs["student_job_match_score"] = (
        (analysis_jobs["skill_overlap_percentage"] * 0.70) +
        (analysis_jobs["tfidf_similarity"] * 100 * 0.30)
    ).round(2)

    top_matching_jobs = analysis_jobs.sort_values("student_job_match_score", ascending=False).head(15).copy()
else:
    analysis_jobs["student_job_match_score"] = []
    top_matching_jobs = pd.DataFrame()

# Add blank score columns to full df for Power BI compatibility
df["matched_skill_count"] = 0
df["matched_skills"] = ""
df["missing_skills_for_job"] = ""
df["skill_overlap_percentage"] = 0.0
df["tfidf_similarity"] = 0.0
df["student_job_match_score"] = 0.0

if not analysis_jobs.empty:
    df.loc[analysis_jobs.index, [
        "matched_skill_count", "matched_skills", "missing_skills_for_job",
        "skill_overlap_percentage", "tfidf_similarity", "student_job_match_score"
    ]] = analysis_jobs[[
        "matched_skill_count", "matched_skills", "missing_skills_for_job",
        "skill_overlap_percentage", "tfidf_similarity", "student_job_match_score"
    ]]


# ------------------------------------------------
# 10. Summaries and TF-IDF
# ------------------------------------------------
role_family_summary = df.groupby(["is_true_data_science_role", "data_role_family"]).agg(
    job_count=("job_key", "nunique"),
    average_skill_count=("skill_count", "mean"),
    average_description_words=("description_word_count", "mean")
).reset_index()

category_summary = df.groupby("category_clean").agg(
    total_jobs=("job_key", "nunique"),
    true_data_science_roles=("is_true_data_science_role", lambda x: (x == "Yes").sum()),
    review_roles=("is_true_data_science_role", lambda x: (x == "Review").sum()),
    average_skill_count=("skill_count", "mean"),
    average_description_words=("description_word_count", "mean")
).reset_index()
category_summary["true_data_science_percentage"] = (
    category_summary["true_data_science_roles"] / category_summary["total_jobs"] * 100
).round(2)
category_summary["category"] = category_summary["category_clean"]
category_summary["number_of_jobs"] = category_summary["total_jobs"]

custom_stop_words = set(ENGLISH_STOP_WORDS).union({
    "sales", "business", "management", "manager", "experience", "work", "team", "company",
    "customer", "support", "job", "role", "employee", "hr", "financial", "development",
    "service", "including", "ability", "skills", "required", "responsibilities",
    "requirements", "year", "years", "knowledge", "strong", "working"
})

if not analysis_jobs.empty:
    try:
        ds_tfidf = TfidfVectorizer(stop_words=list(custom_stop_words), max_features=30, ngram_range=(1, 2))
        ds_tfidf_matrix = ds_tfidf.fit_transform(analysis_jobs["combined_text"])
        ds_tfidf_keywords = pd.DataFrame({
            "keyword": ds_tfidf.get_feature_names_out(),
            "importance_score": ds_tfidf_matrix.sum(axis=0).A1
        }).sort_values("importance_score", ascending=False)
    except Exception as e:
        print("TF-IDF failed:", e)
        ds_tfidf_keywords = pd.DataFrame(columns=["keyword", "importance_score"])
else:
    ds_tfidf_keywords = pd.DataFrame(columns=["keyword", "importance_score"])

kpi_summary = pd.DataFrame({
    "metric": [
        "Total Job Posts",
        "True Data Science / Analytics Roles",
        "Manual Review Roles",
        "True Data Science / Analytics Internships",
        "Target Role Percentage",
        "Required Skills Considered",
        "Matched Skills",
        "Missing Skills",
        "Skill Match Percentage"
    ],
    "value": [
        df["job_key"].nunique(),
        len(analysis_jobs),
        len(review_jobs),
        len(true_ds_internships),
        round((len(analysis_jobs) / df["job_key"].nunique()) * 100, 2),
        len(required_skills),
        len(matched_skills),
        len(missing_skills),
        match_percentage
    ]
})


# ------------------------------------------------
# 11. Create export tables
# ------------------------------------------------
jobs = df[[
    "job_key", "job_id", "category", "category_clean", "job_title",
    "skill_count", "description_word_count", "is_true_data_science_role",
    "data_role_family", "classification_reason", "seniority_level", "is_internship_role",
    "student_job_match_score"
]].copy()

data_science_jobs = analysis_jobs[[
    "job_key", "job_id", "category", "category_clean", "job_title", "data_role_family",
    "seniority_level", "is_internship_role", "skill_count", "description_word_count",
    "matched_skill_count", "matched_skills", "missing_skills_for_job",
    "skill_overlap_percentage", "tfidf_similarity", "student_job_match_score"
]].copy() if not analysis_jobs.empty else pd.DataFrame(columns=[
    "job_key", "job_id", "category", "category_clean", "job_title", "data_role_family",
    "seniority_level", "is_internship_role", "skill_count", "description_word_count",
    "matched_skill_count", "matched_skills", "missing_skills_for_job",
    "skill_overlap_percentage", "tfidf_similarity", "student_job_match_score"
])

review_jobs_export = review_jobs[[
    "job_key", "job_id", "category", "category_clean", "job_title", "data_role_family",
    "classification_reason", "skill_count", "description_word_count"
]].copy() if not review_jobs.empty else pd.DataFrame(columns=[
    "job_key", "job_id", "category", "category_clean", "job_title", "data_role_family",
    "classification_reason", "skill_count", "description_word_count"
])

top_matching_jobs_export = top_matching_jobs[[
    "job_key", "job_id", "category", "category_clean", "job_title", "data_role_family",
    "seniority_level", "is_internship_role", "skill_count", "matched_skill_count",
    "matched_skills", "missing_skills_for_job", "skill_overlap_percentage", "student_job_match_score"
]].copy() if not top_matching_jobs.empty else pd.DataFrame(columns=[
    "job_key", "job_id", "category", "category_clean", "job_title", "data_role_family",
    "seniority_level", "is_internship_role", "skill_count", "matched_skill_count",
    "matched_skills", "missing_skills_for_job", "skill_overlap_percentage", "student_job_match_score"
])

student_profile = pd.DataFrame({
    "student_skill": student_skills,
    "source": "Manual input"
})


# ------------------------------------------------
# 12. Save CSV and Excel files
# ------------------------------------------------
tables = {
    "Jobs": jobs,
    "DataScienceJobs": data_science_jobs,
    "ReviewJobs": review_jobs_export,
    "JobSkills": all_job_skills,
    "DataScienceJobSkills": ds_job_skills,
    "CategorySummary": category_summary,
    "RoleFamilySummary": role_family_summary,
    "TopDataScienceSkills": top_ds_skills,
    "SkillGap": skill_gap,
    "SkillGapSummary": skill_gap_summary,
    "Recommendations": recommendations,
    "TopMatchingJobs": top_matching_jobs_export,
    "DataScienceTFIDFKeywords": ds_tfidf_keywords,
    "StudentProfile": student_profile,
    "KPISummary": kpi_summary
}

for table_name, table_df in tables.items():
    csv_path = powerbi_folder / f"{table_name}.csv"
    table_df.to_csv(csv_path, index=False, encoding="utf-8-sig")
    print("Saved:", csv_path.name)

excel_path = powerbi_folder / "InternBridge_DS_Final_Data.xlsx"
with pd.ExcelWriter(excel_path, engine="openpyxl") as writer:
    for sheet_name, table_df in tables.items():
        table_df.to_excel(writer, sheet_name=sheet_name[:31], index=False)

# Save cleaned dataset with classifications
cleaned_path = outputs_folder / "cleaned_all_job_post_with_ds_roles.csv"
df.to_csv(cleaned_path, index=False, encoding="utf-8-sig")


# ------------------------------------------------
# 13. Save simple report
# ------------------------------------------------
report_text = f"""
InternBridge: Data Science Internship Skill Gap Analyzer
Improved Role Filtering Report

Dataset Summary
---------------
Total job posts: {df['job_key'].nunique()}
True data science / analytics roles: {len(analysis_jobs)}
Manual review roles: {len(review_jobs)}
True data science / analytics internships: {len(true_ds_internships)}
Target role percentage: {round((len(analysis_jobs) / df['job_key'].nunique()) * 100, 2)}%

Why this fix was needed
-----------------------
The previous logic treated many general HR, Finance, Sales, and IT jobs as data science jobs because they mentioned words such as data, Excel, reporting, SQL, dashboard, or analytics.
The new logic only analyzes jobs with strong data science, data analytics, data engineering, machine learning, AI, or business intelligence role evidence.
Ambiguous jobs are marked as Review and are not included in the final skill-gap calculation.

Student Skills
--------------
{', '.join(student_skills)}

Skill Gap Result
----------------
Required skills considered: {len(required_skills)}
Matched skills: {', '.join(matched_skills) if matched_skills else 'None'}
Missing skills: {', '.join(missing_skills) if missing_skills else 'None'}
Skill match percentage: {match_percentage}%

Important Note
--------------
If the goal is strictly data science internships, this dataset currently has very few or no true data science internship roles.
For a stronger internship-focused project, add more internship postings with titles such as Data Science Intern, Data Analyst Intern, Machine Learning Intern, BI Intern, or Data Engineering Intern.
"""

report_path = reports_folder / "InternBridge_DS_Role_Filtering_Report.txt"
with open(report_path, "w", encoding="utf-8") as file:
    file.write(report_text)

print("================================================")
print("IMPROVED INTERNBRIDGE EXPORT COMPLETED")
print("CSV/Excel files saved in:", powerbi_folder)
print("Cleaned dataset saved to:", cleaned_path)
print("Report saved to:", report_path)
print("True data science / analytics roles:", len(analysis_jobs))
print("True data science / analytics internships:", len(true_ds_internships))
print("Manual review roles:", len(review_jobs))
print("Skill match percentage:", match_percentage, "%")
print("================================================")
```
